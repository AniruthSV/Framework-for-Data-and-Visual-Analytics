```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- Personal Info Widgets ---
name = widgets.Text(
    description="Name:",
    placeholder="Enter your name"
)

age = widgets.IntSlider(
    description="Age:",
    min=0, max=100, value=25
)

gender = widgets.ToggleButtons(
    options=['Male', 'Female', 'Other'],
    description='Gender:'
)

birthdate = widgets.DatePicker(
    description='DOB:'
)

height = widgets.FloatSlider(
    description="Height (m):",
    min=1.0, max=2.5, step=0.01, value=1.70
)

bio = widgets.Textarea(
    description="Bio:",
    placeholder="Write something about yourself"
)

# --- Output display area ---
profile_output = widgets.Output()

# --- Submit button ---
submit_btn = widgets.Button(
    description="Create Profile",
    button_style='success',  # 'success', 'info', 'warning', 'danger'
    icon='check'
)

# --- Event handler function ---
def on_submit(b):
    with profile_output:
        clear_output()
        print("ðŸ“„ Profile Summary\n")
        print(f"Name: {name.value}")
        print(f"Age: {age.value}")
        print(f"Height: {height.value} m")
        print(f"Gender: {gender.value}")
        print(f"Date of Birth: {birthdate.value}")
        print(f"Bio: {bio.value}")

# Link the button click event
submit_btn.on_click(on_submit)

# --- Arrange all widgets vertically ---
form = widgets.VBox([
    name,
    age,
    height,
    gender,
    birthdate,
    bio,
    submit_btn,
    profile_output
])

# Display the full form
display(form)
```
